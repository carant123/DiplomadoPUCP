# Renombramos archivos de la carpeta Mask

```
path_root = '/content/drive/MyDrive/Colab Notebooks/mask/'
```

```
# https://electrocooking.wordpress.com/2016/06/17/como-renombrar-ficheros-masivamente-con-python/
nombres = []  #creamos una variable listado vacia para guardar los nombres
#numero = []   #creamos una variable listado vacia para guardar el numero de las fotos
for item in os.listdir(path_root):
    nombres.append(item) #guardamos cada nombre de archivo del directo destino en esta variable

print("Nombre Original: ",nombres)  #mostramos por patalla los nombres originales
print(len(nombres))
```

```
numerodestino = [] # guardamos el numero de destino
cambio = []		   # guardamos el nombre de destino

for index,i in enumerate(range(1001,2479)):  #index es el numero de registro y i es el numero del rango 1014 a 1502
    numerodestino.append(i) # guardamos en un array todos los numeros de destino
    #print("numerodestino: ", numerodestino)
    #rint(range(1000,1010))
    if nombres[index][-4:] == ".jpg": 
        cambio.append("m_" + str(numerodestino[index]) + ".jpg") 
    if nombres[index][-4:] == ".JPG":  
        cambio.append("m_" + str(numerodestino[index]) + ".jpg")  
    if nombres[index][-4:] == ".png":  
        cambio.append("m_" + str(numerodestino[index]) + ".png") 

    os.rename(path_root+nombres[index], path_root+cambio[index])  #aqui ponemos los nombre originales y los nombres destino, ejemplo pasa la imagen DSC_0001.JPG a este nombre DSC_1014.JPG

    #print("original: ", nombres[index])  #mostramos por pantalla los nombre originales
    #print("cambio", cambio[index])   #mostramos por pantalla los nombres de destino
    #print("extension", nombres[index][-4:])  #nos muestra la extension del archivo original
```

```
print("Nombre Cambiado",cambio)
```

```
import cv2
import os
#import glob
#from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```
print(len(os.listdir(path_root)))
```

```
width = 160
height = 160
dsize = (width, height)

for img in os.listdir(path_root): 
    #print(path_root+img)
    src = cv2.imread(path_root+img, cv2.IMREAD_UNCHANGED)
    output = cv2.resize(src, dsize, interpolation = cv2.INTER_AREA)
    cv2.imwrite(path_root_out+img, output)
    
    #plt.figure (figsize = (2,2))
    #plt.imshow (output)
    #plt.axis ('off')
    #plt.show ()
    
    print(path_root_out+img)
```
